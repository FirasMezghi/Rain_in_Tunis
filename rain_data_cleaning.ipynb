```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/drive/MyDrive/dataset ml/rain_csv.csv')
data.head(35)
```

```python
len(data)
```

```python
data = data.loc[data["Day"] != 'Monthly means and totals:']
```

```python
len(data)
```

```python
data.head(35)
```

```python
from numpy.core.numeric import NaN
data['RA']=data['RA'].replace({NaN:0,'o':1})
```

```python
data.head()
```

```python
data['SN']=data['SN'].replace({NaN:0,'o':1})
data['TS']=data['TS'].replace({NaN:0,'o':1})
data['FG']=data['FG'].replace({NaN:0,'o':1})
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.loc[data["Day"] != NaN]
```

```python
data.info()
```

```python
data=data.dropna()
```

```python
data.info()
```

```python
print(data[data=='-'].count())
```

```python
data=data.drop(['SLP','VG','Unnamed: 0'],axis=1)
```

```python
for col in data.columns:
  data[col]=data[col].replace({'-':NaN})
  data = data.fillna(data[col].value_counts().index[0])
```

```python
print(data[data=='-'].count())
```

```python
data.head()
```

```python
data.info()
```

```python
l1=['T', 'TM','Tm','H','PP','VV','V','VM']
l2=['Day','RA','SN','TS','FG']
for i in l1:
  data[i] = data[i].astype(float)
for i in l2:
  data[i] = data[i].astype('int32')
```

```python
data.info()
```

```python
rain_tomorrow=[i for i in list(data['RA'])[1:]]
len(rain_tomorrow)
```

```python
rain_tomorrow.append(0)
data['RA_tomor']=rain_tomorrow
data=data[:-1]
```

```python
data.head()
```

```python
data.info()
```

```python
for feature in l1:
    data=data.copy()
    sns.distplot(data[feature])
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.figure(figsize=(15,15))
    plt.show()
```

```python
data.to_csv('rain_in_tunis_prepared.csv',index=False)
```
